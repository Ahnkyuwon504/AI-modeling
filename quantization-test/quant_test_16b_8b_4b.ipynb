# 2. 모델 추론 테스트

## 2.1. 토크나이저 로드

```python
model_name = "google/gemma-2b-it"

tokenizer = AutoTokenizer.from_pretrained(model_name)
```

## 2.2. 양자화 옵션

```python
bnb_config_8bit = BitsAndBytesConfig(load_in_8bit=True)
bnb_config_4bit = BitsAndBytesConfig(load_in_4bit=True)
```

## 2.3 모델 로드

```python
# 16비트 모델 로드
model_16bit = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.float16)

# 8비트 모델 로드
model_8bit = AutoModelForCausalLM.from_pretrained(model_name, quantization_config=bnb_config_8bit)

# 4비트 모델 로드
model_4bit = AutoModelForCausalLM.from_pretrained(model_name, quantization_config=bnb_config_4bit)
```

```python
model_16bit.save_pretrained('model-16bit')
model_8bit.save_pretrained('model-8bit')
model_4bit.save_pretrained('model-4bit')
```

# 3. 모델 추론 실험

## 3-1. 성능 측정 함수

```python
input = """
1호선에서 9호선으로 갈아타고 다시 4호선으로 환승하면 뱅크웨어글로벌에 출근할 수 있어.
이 때 지하철을 몇 번 갈아탔을까?
"""
```

```python
def measure_performance(model, curr):
    start_time = time()

    messages = [
        {
            "role": "user",
            "content": """
            다음 질문에 대답해 주세요.\n\n {}
            """.format(curr)
        }
    ]

    pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=128)
    prompt = pipe.tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)

    inputs = tokenizer.encode(prompt, add_special_tokens=False, return_tensors="pt")
    outputs = model.generate(input_ids=inputs.to(model.device), max_new_tokens=150)

    end_time = time()
    return tokenizer.decode(outputs[0]), end_time - start_time
```

## 3-2. 실험

```python
# 16비트 모델 성능 측정
output_16bit, time_16bit = measure_performance(model_16bit, input)
print(f"16비트 모델 출력: {output_16bit}")
print(f"16비트 모델 추론 시간: {time_16bit:.4f}초")
```

```python
# 8비트 모델 성능 측정
output_8bit, time_8bit = measure_performance(model_8bit, input)
print(f"8비트 모델 출력: {output_8bit}")
print(f"8비트 모델 추론 시간: {time_8bit:.4f}초")
```

```python
# 4비트 모델 성능 측정
output_4bit, time_4bit = measure_performance(model_4bit, input)
print(f"4비트 모델 출력: {output_4bit}")
print(f"4비트 모델 추론 시간: {time_4bit:.4f}초")
```
